```python
# Your code here
import requests
import zipfile
import io
import os
import pandas as pd


def download_census_data(year, code, base_dir=''):
  year = str(year)
  url = f'https://www.nomisweb.co.uk/output/census/{year}/census{year}-{code.lower()}.zip'
  extract_dir = os.path.join(base_dir, os.path.splitext(os.path.basename(url))[0])

  if os.path.exists(extract_dir) and os.listdir(extract_dir):
    print(f"Files already exist at: {extract_dir}.")
    return

  os.makedirs(extract_dir, exist_ok=True)
  response = requests.get(url)
  response.raise_for_status()

  with zipfile.ZipFile(io.BytesIO(response.content)) as zip_ref:
    zip_ref.extractall(extract_dir)

  print(f"Files extracted to: {extract_dir}")

def load_census_data(year, code, level='msoa'):
  return pd.read_csv(f'census{year}-{code.lower()}/census{year}-{code.lower()}-{level}.csv')
```

```python
uk_min_lon, uk_max_lon = -10.0, 2.0
uk_min_lat, uk_max_lat = 49.0, 61.0
tile_bounding_boxes = {}
tile_counter = 0

for lon in range(int(uk_min_lon), int(uk_max_lon)):
    for lat in range(int(uk_min_lat), int(uk_max_lat)):
        tile_bounding_boxes[tile_counter] = (lon, lon + 1, lat, lat + 1)
        tile_counter += 1
```

```python
final_tiles = []
for k, (w, e, s, n) in tile_bounding_boxes.items():
  if w > max_lon or e < min_lon or s > max_lat or n < min_lat:
    continue
  final_tiles.append(k)
```

```python
import osmium

class OSMHandler(osmium.SimpleHandler):
    def __init__(self):
        super().__init__()
        self.nodes = []
        self.ways = []
        self.relations = []

    def node(self, n):
        if n.tags:
          self.nodes.append({
              'id': n.id,
              'lat': n.location.lat,
              'lon': n.location.lon,
              'tags': dict(n.tags)
          })

    def way(self, w):
        if w.tags:
          self.ways.append({
              'id': w.id,
              'nodes': [n.ref for n in w.nodes],
              'tags': dict(w.tags)
          })

    def relation(self, r):
        if r.tags:
          self.relations.append({
              'id': r.id,
              'members': [(m.ref, m.role) for m in r.members],
              'tags': dict(r.tags)
          })

handler = OSMHandler()
```

```python
import json

for i in final_tiles:
    handler.apply_file(f"tile_{i}.osm.pbf")
    nodes_df = pd.DataFrame(handler.nodes)
    nodes_df['tags'] = nodes_df['tags'].apply(json.dumps)
    nodes_df = nodes_df.drop_duplicates()
    nodes_df.to_sql('osm_nodes', con=engine, if_exists='append', index=False)
    print("Added osm tile " + str(i) + " to the database")
```

```python
def string_to_dict(tag_string):
    tag_string = tag_string[1:-1]
    if pd.isna(tag_string):
        return None
    try:
        return {k[1:-1]: v[1:-1] for k, v in (item.split(': ') for item in tag_string.split(', '))}
    except ValueError:
        return None
```

```python
def get_pois(latitude, longitude):
  box = 0.02
  south, north = latitude-box, latitude+box
  west, east = longitude-box, longitude+box
  pois = %sql SELECT DISTINCT * FROM osm_nodes WHERE lat between :south AND :north AND lon BETWEEN :west AND :east;
  pois_df = pois.DataFrame()
  return pois_df
```

```python
def tags_counter(pois_df):
  counter = {}
  features = pois_df['tags']

  for i in range(len(features)):
    feature = string_to_dict(features[i])
    try:
      for k, v in feature.items():
        if k not in counter:
          counter[k] = {}
        if v not in counter[k]:
          counter[k][v] = 0
        counter[k][v] += 1
    except:
      continue

  return counter
```

```python
cambridge_counter = tags_counter(cambridge_pois)
print(cambridge_counter)
```

```python
def features_over_n(counter, n):
    top_feat = {}
    for k, v in counter.items():
      v_keys = list(v.keys())
      for key in v_keys:
        if v[key] > n:
          if k not in top_feat:
            top_feat[k] = []
          top_feat[k].append((key, v[key]))

    return top_feat
```

```python
from collections import Counter

def count_keys(tags, keys):
    return Counter({key: tags.get(key, None) is not None for key in keys})

def count_specific_keys(tags, specific_keys):
    tag_counter = Counter()

    for tag in specific_keys:
        key, value = tag
        if tags.get(key) == value:
            tag_counter[f'{key}={value}'] = 1
        else:
            tag_counter[f'{key}={value}'] = 0

    return tag_counter
```

```python
from copy import Error
def extract_tag_counts(tags_df, keys, specific_keys):
  try:
    tags_df["tags"] = tags_df["tags"].apply(lambda x: string_to_dict(x))
    tags_df = tags_df.dropna()
  except:
    raise Error

  key_counts = tags_df['tags'].apply(lambda x: count_keys(x, keys))
  specific_key_counts = tags_df['tags'].apply(lambda x: count_specific_keys(x, specific_keys))
  total_key_counts = sum(key_counts, Counter())
  total_specific_key_counts = sum(specific_key_counts, Counter())

  key_count_df = pd.DataFrame(total_key_counts.items(), columns=['Key', 'Count'])
  specific_key_count_df = pd.DataFrame(total_specific_key_counts.items(), columns=['Key', 'Count'])
  count_df = pd.concat([key_count_df, specific_key_count_df], axis=0, ignore_index=True)

  for k in keys:
    if k not in list(count_df["Key"]):
      count_df.loc[len(count_df)] = [k, 0]
  for k1, k2 in specific_keys:
    col = k1 + "=" + k2
    if col not in list(count_df["Key"]):
      count_df.loc[len(count_df)] = [col, 0]

  return count_df

def change_count_df_format(count_df, key):
  count_df.loc[len(count_df)] = ["OA21CD", key]
  df2 = pd.DataFrame(columns=count_df["Key"])
  df2.loc[0] = count_df['Count'].values

  return df2
```
